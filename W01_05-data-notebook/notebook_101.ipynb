## Jupyter Notebook nedir?

Notebook, hem kod (örneğin Python) hem de zengin metin öğeleri (paragraflar, denklemler, şekiller, linkler vb.) içeren bir dokümandır; bu zengin içerik Markdown [Markdown](https://guides.github.com/features/mastering-markdown/) ve MathJax [MathJax](https://www.mathjax.org/) sayesinde mümkün olur.

Jupyter Notebook üç bileşeni bir araya getirir:

1. **Notebook web uygulaması**:Kod yazmak ve çalıştırmak, ayrıca notebook dokümanları oluşturmak için kullanılan etkileşimli bir web uygulamasıdır.
2. **Kernel’ler** :Notebook web uygulaması tarafından başlatılan ve kullanıcının kodunu belirli bir dilde çalıştırıp çıktıyı tekrar notebook web uygulamasına geri gönderen ayrı process’lerdir. Kernel aynı zamanda interactive widget’lar için yapılan hesaplamalar, tab tamamlama (tab completion) ve introspection gibi işleri de yürütür.
3. **Notebook dokümanları**: Notebook web uygulamasında gördüğün tüm içeriğin bir temsilini içinde barındıran, kendi kendine yeterli dokümanlardır; hesaplamaların input ve output’ları, açıklayıcı metinler, denklemler, görseller ve objelerin zengin medya gösterimleri buna dahildir. Her notebook dokümanının kendi kernel’i vardır.

## Kernels

Jupyter’ın kernel ve messaging mimarisi sayesinde Notebook, birçok farklı programlama dilinde kod çalıştırabilir. Bir kullanıcı bir notebook dokümanı açtığında, web uygulaması o notebook için kodu çalıştıracak bir kernel başlatır. Her kernel tek bir programlama dilinde kod çalıştırabilir ve şu diller için mevcut kernel’ler vardır:

- [Python](https://github.com/ipython/ipython)
- [Julia](https://github.com/JuliaLang/IJulia.jl)
- [R](https://github.com/IRkernel/IRkernel)
- [Ruby](https://github.com/sciruby/iruby)

Varsayılan kernel Python’dur. Notebook, kullanıcıların hangi notebook’un hangi kernel ile çalışacağını kolayca seçmesini sağlar.

Bu kernel’lerin her biri, notebook web uygulaması ve web tarayıcısı ile JSON tabanlı bir protokol üzerinden ZeroMQ/WebSockets kullanarak iletişim kurar. Bu protokolün detayları burada [here](https://jupyter-client.readthedocs.io/en/stable/messaging.html) açıklanmıştır. Kullanıcıların çoğunun bu ayrıntıları bilmesine gerek yoktur; ama “kernel’ler kodu çalıştırır” mantığını anlamak önemlidir.


Eğer bilgisayarında hangi kernel’lerin yüklü olduğunu hatırlamıyorsan, terminalde şu komutu çalıştırabilirsin:

```bash
jupyter kernelspec list
```

## Belgeler

Notebook belgeleri, çalıştırılabilir bir oturumun giriş ve çıkışlarını içerir; ayrıca koda eşlik eden fakat yürütülmesi amaçlanmayan açıklayıcı metinleri de barındırır. Kod çalıştırıldığında oluşan zengin çıktılar—HTML, görseller, videolar ve grafikler—doğrudan notebook içine gömülür. Bu da notebook’u bir hesaplamanın tam ve kendi içinde yeterli bir kaydı hâline getirir.

Notebook web uygulamasını bilgisayarında çalıştırdığında, notebook belgeleri yerel dosya sisteminde .ipynb uzantılı dosyalardır. Böylece bu dosyaları klasörlere ayırma, düzenleme ve başkalarıyla paylaşma gibi alışık olduğun iş akışlarını kullanabilirsin.

## Editor

Jupyter Notebook modlu (modal) bir kullanıcı arayüzüne sahiptir. Bu, klavyenin işlevinin Notebook’un hangi modda olduğuna bağlı olarak değiştiği anlamına gelir. İki mod vardır: edit (düzenleme) modu ve command (komut) modu.
    
### Edit mode (Düzenleme Modu)

Edit mode, hücre giriş düzenleme alanının etrafındaki mavi dar bir çerçeve ve yanıp sönen imleç ile belirtilir (ince yanıp sönen dikey çizgi):

![Cell in edit mode with tight border](https://wagon-public-datasets.s3.amazonaws.com/data-science-images/02-Data-Toolkit/01-Data-Analysis/jupyter-edit-mode.png)

Bir hücre edit modundayken, hücreye normal bir metin düzenleyici gibi yazabilirsiniz.

Edit moduna girmek için `Enter` tuşuna basın veya fareyle hücrenin düzenleme alanına tıklayın. 

### Command mode (Komut modu)

Command mode, hücrenin tamamının etrafındaki daha geniş bir mavi çerçeve ile belirtilir:

![Cell in command mode with border further away](https://wagon-public-datasets.s3.amazonaws.com/data-science-images/02-Data-Toolkit/01-Data-Analysis/jupyter-command-mode.png)

Komut modundayken, hücrelerin içeriğine yazamazsınız; bunun yerine notebook’un bütününü düzenleyebilirsiniz. Bu modda klavye, hücre ve notebook işlemlerini hızlıca yapabilmenizi sağlayan kısayollarla eşleştirilmiştir. Örneğin, komut modundayken C tuşuna basarsanız geçerli hücreyi kopyalarsınız.

Komut modundayken hücre içine yazmaya çalışmayın: beklenmedik şeyler olabilir!

Komut moduna girmek için Esc tuşuna basın veya fareyle hücrenin düzenleme alanının dışına tıklayın.


## Klavye Kısayolları

Klavye kısayollarını öğrenmek, herhangi bir notebook’u daha hızlı yönetebilmek için çok önemlidir. İstediğiniz anda tüm kısayolları içeren bir pencere açmak için araç çubuğundaki Help menüsüne, ardından Keyboard Shortcuts seçeneğine tıklayabilirsiniz. Ayrıca komut modundayken klavyenizde H tuşuna basabilirsiniz.

Bugün öğrenmeye başlamanızı önerdiğimiz birkaç kısayol:

**Command** (komut) modundayken:
- `A` Geçerli hücrenin üstüne **Above** bir hücre ekler
- `B` Geçerli hücrenin altına **Below** bir hücre ekler
- `D`, `D` (iki kez `D` ) Geçerli hücreyi siler. Kurtarmak için `Edit` > `Undo Delete Cells`
- `Y` Geçerli hücre türünü **Code** (Python) yapar
- `M` Geçerli hücre türünü **Markdown** yapar
- `↩` ((`Enter` ) veya çift tıklama ) **edit**  moduna girer
- `↑`/`↓` ((Yukarı / Aşağı ok) `Up`/`Down` ) Hücreler arasında gezinir

 **edit** modundayken:

- `Esc` Komut moduna geçer (yani edit modundan çıkar)
- `Ctrl`-`↩` ( `Ctrl` `Enter`) Kodu çalıştırır ve aynı hücrede komut moduna geçer
- `⇧`-`↩` ( `Shift` `Enter`) Kodu çalıştırır, altına yeni bir hücre oluşturur ve o yeni hücrede edit moduna girer

## [Opsiyonal] Matematik

Jupyter Notebook’un içindeki Markdown işleyicisi MathJax (**MathJax-aware**) desteklidir. Bu da MathJax’in Tex ve LaTeX alt kümesini kullanarak matematiksel ifadeleri özgürce yazabileceğin anlamına gelir. 

Bir cümlenin içine matematiksel ifade eklemek için `$`kullanabilirsin. Örneğin,  `$\sqrt{3x-1}+(1+x)^2$` şu şekilde görüntülenir: $\sqrt{3x-1}+(1+x)^2$.

Aşağıdaki Markdown hücrelerini çalıştırmayı dene ve sonucu keyifle izle! (Source: [Motivating Examples - Jupyter Notebook](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Typesetting%20Equations.html))

### (`n`) madeni para atıldığında (`k`) kez tura gelme olasılığı

\begin{equation*}
P(E) = {n \choose k} p^k (1-p)^{ n-k}
\end{equation*}

### Fizikte Maxwell Denklemleri

\begin{align}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\   \nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0
\end{align}
